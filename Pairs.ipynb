```python
import itertools
import random

import jupyter_black
from tabulate import tabulate

jupyter_black.load()

# comment out people who are not present
# or prefer to work alone

group1 = [
    "Julia H",
    "Jannis B",
    # "Maxime D",
    "Louisa R",
    # "Klara D",
    "Quirin S",
    "Johanna S",
    "Marlene S",
    "Maximilian E",
    "Luis B",
]
group2 = [
    "Fabian W",
    # "Lukas N",
    "Moritz M",
    "Sara B",
    "Antoaneta M",
    "Julian W",
    # "Clemens P",
    "Marla P",
    "Nicole W",
]

# keep this list up-to-date
prev_pairs = [
    [
        ("Johanna S", "Luis B"),
        ("Jannis B", "Antoaneta M"),
        ("Maximilian E", "Nicole W"),
        ("Louisa R", "Lukas N"),
        ("Marlene S", "Sara B"),
        ("Quirin S", "Julian W"),
        ("Maxime D", "Fabian W"),
        ("Klara D", "Marla P"),
        ("Julia H", "Moritz M"),
    ],
    [
        ("Klara D", "Antoaneta M"),
        ("Maximilian E", "Moritz M"),
        ("Marlene S", "Luis B"),
        ("Quirin S", "Fabian W"),
        ("Julia H", "Sara B"),
        ("Jannis B", "Marla P"),
        ("Maximilian E", "Sara B"),
        ("Johanna S", "Lukas N"),
        ("Maxime D", "Julian W"),
    ],
    [
        ("Johanna S", "Antoaneta M"),
        ("Jannis B", "Julian W"),
        ("Louisa R", "Nicole W"),
        ("Luis B", "Moritz M"),
        ("Klara D", "Sara B"),
        ("Maxime D", "Marla P"),
        ("Quirin S", "Clemens P"),
        ("Maximilian E", "Fabian W"),
    ],
]


all_pairs = list(itertools.product(group1, group2))
random.shuffle(all_pairs)
paired_students = set()
selected_pairs = []

# Iterate over shuffled pairs and select unique pairs
for pair in all_pairs:
    student1, student2 = pair
    new_pair = True
    # Check if students were already paired in a previous week
    for prev_week in prev_pairs:
        if pair in prev_week:
            new_pair = False
            break
    if not new_pair:
        continue
    # Check if neither student has been paired yet in current week
    if student1 not in paired_students and student2 not in paired_students:
        # Add students to the set of paired students
        paired_students.add(student1)
        paired_students.add(student2)
        # Add the pair to the list of selected pairs
        selected_pairs.append(pair)
    # Check if we have enough pairs
    if len(selected_pairs) == 9:
        break

not_selected = [
    student
    for student in (group1 + group2)
    if student not in [student for pair in selected_pairs for student in pair]
]

# print in org format
print(
    tabulate(
        [[i + 1, *pair] for i, pair in enumerate(selected_pairs)],
        ["Pair", "Student 1", "Student 2"],
        tablefmt="orgtbl",
    )
)
print("\nUnassigned:", not_selected)
```
