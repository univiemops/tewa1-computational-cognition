```python
import itertools
import random

import jupyter_black
from tabulate import tabulate

jupyter_black.load()

# comment out people who are not present
# or prefer to work alone

group1 = [
    "Julia H",
    "Jannis B",
    # "Maxime D",
    # "Louisa R",
    # "Klara D",
    "Quirin S",
    "Johanna S",
    "Marlene S",
    # "Maximilian E",
    "Luis B",
]
group2 = [
    # "Fabian W",
    # "Lukas N",
    "Moritz M",
    # "Sara B",
    # "Antoaneta M",
    "Julian W",
    # "Clemens P",
    # "Marla P",
    "Nicole W",
]

# keep this list up-to-date
prev_pairs = [
    [
        ("Johanna S", "Luis B"),
        ("Jannis B", "Antoaneta M"),
        ("Maximilian E", "Nicole W"),
        ("Louisa R", "Lukas N"),
        ("Marlene S", "Sara B"),
        ("Quirin S", "Julian W"),
        ("Maxime D", "Fabian W"),
        ("Klara D", "Marla P"),
        ("Julia H", "Moritz M"),
    ],
    [
        ("Klara D", "Antoaneta M"),
        ("Maximilian E", "Moritz M"),
        ("Marlene S", "Luis B"),
        ("Quirin S", "Fabian W"),
        ("Julia H", "Sara B"),
        ("Jannis B", "Marla P"),
        ("Maximilian E", "Sara B"),
        ("Johanna S", "Lukas N"),
        ("Maxime D", "Julian W"),
    ],
    [
        ("Johanna S", "Antoaneta M"),
        ("Jannis B", "Julian W"),
        ("Louisa R", "Nicole W"),
        ("Luis B", "Moritz M"),
        ("Klara D", "Sara B"),
        ("Maxime D", "Marla P"),
        ("Quirin S", "Clemens P"),
        ("Maximilian E", "Fabian W"),
    ],
    [
        ("Julia H", "Julian W"),
        ("Luis B", "Nicole W"),
        ("Jannis B", "Fabian W"),
        ("Quirin S", "Antoaneta M"),
        ("Marlene S", "Marla P"),
        ("Johanna S", "Moritz M"),
        ("Louisa R", "Sara B"),
        ("Lukas N", "Maximilian E"),
    ],
]
prev_pairs_flat = [a for b in prev_pairs for a in b]

all_pairs = list(itertools.product(group1, group2))
random.shuffle(all_pairs)
paired_students = set()
selected_pairs = []

# Iterate over shuffled pairs and select unique pairs
for pair in all_pairs:
    student1, student2 = pair
    new_pair = True
    # Check if students were already paired in a previous week
    for prev_week in prev_pairs:
        if pair in prev_week:
            new_pair = False
            break
    if not new_pair:
        continue
    # Check if neither student has been paired yet in current week
    if student1 not in paired_students and student2 not in paired_students:
        # Add students to the set of paired students
        paired_students.add(student1)
        paired_students.add(student2)
        # Add the pair to the list of selected pairs
        selected_pairs.append(pair)
    # Check if we have enough pairs
    if len(selected_pairs) == 9:
        break

not_selected = [
    student
    for student in (group1 + group2)
    if student not in [student for pair in selected_pairs for student in pair]
]

# print in org format
print(
    tabulate(
        [[i + 1, *pair] for i, pair in enumerate(selected_pairs)],
        ["Pair", "Student 1", "Student 2"],
        tablefmt="orgtbl",
    )
)
print("\nUnassigned:", not_selected)
```

Form pairs from both groups

```python
names = [
    "Julia H",
    "Jannis B",
    # "Maxime D",
    # "Louisa R",
    # "Klara D",
    "Quirin S",
    "Johanna S",
    "Marlene S",
    # "Maximilian E",
    "Luis B",
    # "Fabian W",
    "Lukas N",
    "Moritz M",
    # "Sara B",
    # "Antoaneta M",
    "Julian W",
    # "Clemens P",
    # "Marla P",
    "Nicole W",
]


def generate_pairs(names, forbidden_pairs):
    # Create a copy of the names list
    remaining_names = names.copy()

    # Initialize an empty list to store the pairs
    pairs = []

    # Continue pairing until there's only one name left
    while len(remaining_names) > 1:
        # Randomly select the first name
        name1 = random.choice(remaining_names)
        remaining_names.remove(name1)

        # Filter out forbidden pairs based on name1
        filtered_names = [
            name
            for name in remaining_names
            if (name1, name) not in forbidden_pairs
            and (name, name1) not in forbidden_pairs
        ]

        # If there are no valid pairs left for name1, continue to the next iteration
        if not filtered_names:
            continue

        # Randomly select the second name from the filtered names
        name2 = random.choice(filtered_names)
        remaining_names.remove(name2)

        # Add the pair to the pairs list
        pairs.append((name1, name2))

    # If there's one name left, add it to the unpaired list
    if remaining_names:
        pairs.append((remaining_names[0], None))

    return pairs


selected_pairs = generate_pairs(names, prev_pairs_flat)
print("Random pairs:", selected_pairs)
```
